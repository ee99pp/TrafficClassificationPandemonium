```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['SimHei'] # 用来正常显示中文标签SimHei
plt.rcParams['axes.unicode_minus'] = False # 用来正常显示负号

npy_label_path = "../datasets/android/train/label.npy"
# 示例字典与标签数据
label_dict = {'qq': 0, '微信': 1, '淘宝': 2}


```

```python
label_data = np.load(npy_label_path)  # 请替换为你的实际文件路径

# 统计各类别样本数量
label_counts = {label: np.sum(label_data == index) for label, index in label_dict.items()}

# 绘制条形图
fig, ax = plt.subplots()
categories = list(label_counts.keys())
counts = list(label_counts.values())

ax.bar(categories, counts)
ax.set_xlabel('Categories')
ax.set_ylabel('Sample Counts')
ax.set_title('Sample Counts for Each Category')
ax.set_xticklabels(label_counts.keys(), rotation=-45, fontsize=12) # 设置字体大小和倾斜
plt.show()
```
